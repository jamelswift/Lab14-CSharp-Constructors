## Constructors

2.เปลี่ยน code ให้เป็นดังต่อไปนี้

```csharp
var circle = new Circle();
var rectangle = new Rectangle();
var triangle = new Triangle();

class Shape
{
    private int? NumOfSide;
    private Shape()
    {
        System.Console.WriteLine("This is some shape with unknown side");
    }
    public Shape(int NumOfSide)
    {
        System.Console.WriteLine($"This is some shape with {NumOfSide} sides" );
    }
}
class Circle :Shape
{
    public Circle():base()
    {
        System.Console.WriteLine("This is a circle");
    }
}
class Rectangle :Shape
{
    public Rectangle(): base(4)
    {
        System.Console.WriteLine("This is a rectangle");
    }
}
class Triangle :Shape
{
   public Triangle() : base(3)
    {
        System.Console.WriteLine("This is a triangle");
    }
}
```

สรุปการทดลอง:

การใช้ base() เพื่อเรียกคอนสตรัคเตอร์ของคลาสแม่: คอนสตรัคเตอร์ของคลาสลูก (Circle, Rectangle, Triangle) ใช้ base() เพื่อเรียกคอนสตรัคเตอร์ของคลาสแม่ (Shape) ซึ่งช่วยในการกำหนดค่าพื้นฐานหรือข้อมูลเริ่มต้นสำหรับอ็อบเจ็กต์ในคลาสแม่.

การใช้ protected สำหรับคอนสตรัคเตอร์: การเปลี่ยนคอนสตรัคเตอร์ของคลาสแม่จาก private เป็น protected ทำให้สามารถเข้าถึงคอนสตรัคเตอร์นี้จากคลาสลูกได้, ซึ่งในกรณีนี้ช่วยให้คลาสลูกสามารถใช้คอนสตรัคเตอร์ของ Shape ได้อย่างถูกต้อง.

การสืบทอดลักษณะของคอนสตรัคเตอร์: คอนสตรัคเตอร์ที่ใช้ในคลาสลูกช่วยให้การสร้างอ็อบเจ็กต์จากคลาสลูกมีลักษณะที่เกี่ยวข้องกับคลาสแม่ และทำให้สามารถสื่อสารระหว่างคลาสแม่และคลาสลูกได้ดีขึ้น.
